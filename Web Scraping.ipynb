```python
from bs4 import BeautifulSoup
import time
import pandas as pd
from requests_html import HTMLSession
```

```python
session = HTMLSession()
antapani = 'https://www.lamudi.co.id/jual/jawa-barat/bandung/antapani-cicadas/rumah/?sort=terbaru'
buahbatu = 'https://www.lamudi.co.id/jual/jawa-barat/bandung/buahbatu/rumah/?sort=terbaru'
arcamanik = 'https://www.lamudi.co.id/jual/jawa-barat/bandung/arcamanik/rumah/?sort=terbaru'
pagination  = '&page='
```

```python
def get_data(url):
    r = session.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup
```

```python
listinglink = []
listrumah = []
```

```python
for i in range(1,22):
    soup = get_data(arcamanik+pagination+str(i))
    listing = soup.find_all('div', class_='row ListingCell-row ListingCell-agent-redesign')
    for list in listing:
        link = list.find('a', class_="js-listing-link")['href']
        listinglink.append(link)
    print(len(listinglink), 'Link Didapat')
    time.sleep(0.1)
```

```python
for link in listinglink:
    soup = get_data(link)

    try:
        judul = soup.find('h1').text.strip()
    except:
        judul = ""

    kecamatan = "Arcamanik"

    try:
        harga = soup.find('span', class_='FirstPrice').text.strip()
    except:
        harga = ""

    spesifikasi = soup.find_all('div', class_='ellipsis')
    nspesifikasi = soup.find_all('div', class_='last')

    spek = []
    for x in spesifikasi:
        spek.append(x.text.strip())

    try:
                    iparkir = spek.index("Lahan Parkir")
                    parkir = nspesifikasi[iparkir].text.strip()
    except:
                    parkir = ""

    try:
                    itidur = spek.index("Kamar tidur")
                    ktidur = nspesifikasi[itidur].text.strip()
    except:
                    ktidur = ""

    try:
                    imandi = spek.index("Kamar mandi")
                    kmandi = nspesifikasi[imandi].text.strip()
    except:
                    kmandi = ""

    try:
                    ilb = spek.index("Bangunan (m²)")
                    lb = nspesifikasi[ilb].text.strip()
    except:
                    lb = ""

    try:
                    ilt = spek.index("Lahan (m²)")
                    lt = nspesifikasi[ilt].text.strip()
    except:
                    lt = ""
    info_properti = {
                    'Link'          : link,
                    'Judul'         : judul,
                    'Kecamatan'     : kecamatan,
                    'Lahan Parkir'  : parkir,
                    'Kamar Tidur'   : ktidur,
                    'Kamar Mandi'   : kmandi,
                    'Luas Bangunan' : lb,
                    'Luas Tanah'    : lt,
                    'Harga'         : harga
                }
    listrumah.append(info_properti)
    print('Data Ke -',len(listrumah),' : ', info_properti['Judul'])
    time.sleep(0.5)
```

```python
df = pd.DataFrame(listrumah)
print(df.head(5))
```

```python
df.to_csv('Arcamanik.csv', header=True, index=False)
```
